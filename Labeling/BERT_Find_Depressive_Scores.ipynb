```python
import pandas as pd
import transformers
```

```python
df = pd.read_csv('clean_tweet_Dec19ToDec20_normalized.csv')
df.info()
```

```python
df.tail(10)
```

```python
dataset = df['text']
```

```python
dataset.info()
```

```python
classifier = transformers.pipeline("zero-shot-classification", model="digitalepidemiologylab/covid-twitter-bert-v2-mnli")
```

```python
labels = sorted(['Depressive', 'Not Depressive'], reverse=False)
labels
```

```python
labeled_dataset = []
```

```python
for text in dataset:
    result = classifier(text, labels, multi_label=False)
    text = result['sequence']
    if (labels == result['labels']) :
      labelss = result['labels']
      scores = result['scores']
    else :
      labelss = sorted(result['labels'], reverse=False)
      scores = sorted(result['scores'], reverse=False)
    labeled_dataset.append((text, labelss, scores))
```

```python
labeled_dataset
```

```python
df_labeled = pd.DataFrame(labeled_dataset, columns=['sequence', 'labels', 'scores'])
```

```python
df_labeled.head(-10)
```

```python
df_labeled.scores[0][0]
```

```python
df_labeled.info()
```

```python
df_labeled.head(10)
```

```python
df_depression = pd.DataFrame(df_labeled, columns=['sequence', 'labels', 'scores'])
```

```python
df_depression.scores[0][0]
```

```python
def separate_scores(row):
    if row['scores'] is not None:
        return pd.Series({'Depressive': row['scores'][0], 'Not Depressive': row['scores'][1]})
    return pd.Series({'Depressive': None, 'Not Depressive': None})


df_depression[['depression', 'not_depression']] = df_depression.apply(separate_scores, axis=1)
```

```python
df_depression.to_csv("twitter_llm_depressive_clean_full.csv", index=False)
```
