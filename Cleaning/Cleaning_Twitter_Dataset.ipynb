```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
import re
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

stopWords = set(stopwords.words('english'))

def tweet_cleaner(text):

    #remove RT
    text = re.sub("RT @[\w]*:","",text)
    #remove twitter handles (@user)
    text = re.sub("(@[A-Za-z0-9_]+)","", text)
    #remove url links
    text = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+#]|[!*\(\),]|'\
                  '(?:%[0-9a-fA-F][0-9a-fA-F]))+','', text)
    text = text.lower()

    # removing stop-words
    words = text.split()
    filter_text = [word for word in words if word not in stopWords]

    text = ' '.join(filter_text)

    return text
```

```python
#Apply text cleaning function to cleaned_text column in df_tweets dataframe
df['text'] = df['text'].apply(lambda x: tweet_cleaner(x))
```
